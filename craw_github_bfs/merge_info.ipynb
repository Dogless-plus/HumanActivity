```python
from os import listdir,sep
import pandas as pd
```

```python
dirs = listdir("info")
dirs[:5]
```

```python
with open("info"+sep+"0.csx","rt") as f:
    data = f.readline()
data
```

```python
rows = []
for userfile in dirs:
    user = userfile.split(".")[0]
    with open("info"+sep+userfile,"rt") as f:
        data = f.readline().split("|||")
    datarow=(user,*data)
    rows.append(datarow)
#     print(datarow)
```

```python
df = pd.DataFrame(rows,columns=["id","nfollower","nfollowing","location","ngist","nrepo"])
```

```python
df.to_csv("info.csv",encoding="utf-8",index=False)
```
