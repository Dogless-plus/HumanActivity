```python
starts,ends = [],[]
count = 0
for userfile in dirs:
    count+=1
    if not count%30000:
        print(count)
    with open("trace"+sep+userfile ,"rt") as f:
        lines = f.readlines()
    starts.append(lines[0].strip().split(",")[-1])
    ends.append(lines[-1].strip().split(",")[-1])
```

```python
min(starts),max(starts),min(ends),max(ends)
```

```python
df = pd.read_csv("trace"+sep+"0.csv",encoding="utf-8",header=None,names=["value","day"])
df["day"] = pd.to_datetime(df["day"])
```

```python
values = list(df[(df["day"] > pd.to_datetime("2016-04-16")) & (df["day"] < pd.to_datetime("2017-04-15"))]["value"])  # 2016-4-17(7) -> 2016-4-14
```

```python
len(values)
```

```python
trace_data= []
count =0 
for userfile in dirs:
    count+=1
    if not count%10000:
        print(count)
    user=userfile.split(".")[0]
    df = pd.read_csv("trace"+sep+userfile,encoding="utf-8",header=None,names=["value","day"])
    df["day"] = pd.to_datetime(df["day"])
    values = list(df[(df["day"] > pd.to_datetime("2016-04-16")) & (df["day"] < pd.to_datetime("2017-04-15"))]["value"])  # 2016-4-17(7) -> 2016-4-14
    datarow = (user,*values)
    trace_data.append(datarow)
```

```python
df_trace = pd.DataFrame(trace_data)
```

```python
df_trace.to_csv("trace.csv",encoding="utf-8",header= None, index=None)
```
