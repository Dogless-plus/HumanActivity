```python
df_info = pd.read_csv("info.csv",header=0,encoding="utf-8")
df_info.head()
```

```python
df_trace = pd.read_csv("trace.csv",header=None,encoding="utf-8",names=["d_"+str(i) for i in range(363)])
df_trace.reset_index(inplace=True)
df_trace.head()
```

```python
df_info.shape,df_trace.shape
```

```python
df_merge=df_info[["id","nfollower","nfollowing","ngist","nrepo"]].merge(
    df_trace,how="inner",left_on="id",right_on="index").drop(["index",],axis=1)
df_merge.head()
```

```python
df_merge.to_csv("data_info_trace.csv",encoding="utf-8",index=False)
```
